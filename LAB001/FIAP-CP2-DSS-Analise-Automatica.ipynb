### Projeto: Análise de dados automatizada

CP 2 - 2 Semestre

> Tarefa no MS Teams, link para o repo no GitHub com o Jupyter notebook do projeto, e exemplo de HTML com análise gerado

No repo do GitHub
- Jupyter notebook com o código do projeto
- Arquivo .html com análise de data frame Pandas realizado de forma automatizada (testes)

- Equipes de 2 ou 3 alunos
  - Projeto não pode ser individual
  - Equipe não pode ser > 3 alunos por grupo
- Escopo do Projeto: Criar uma ferramenta para análise de dados automatizadad
  - Criar uma função que receba um data frame Pandas como parametro
  - O data frame deve ser analisado de forma automatizada
  - O resultado da análise deve ser salva (exportada), deve gerar um arquivo .html auto contido
  - A análise deve gerar gráficos, como do Seaborn por exemplo
  - Mostrar gráfico de correlações entre as features (séries, colunas do data frame)
  - Analisar os dados de cada coluna do data frame
  - Realizar regressão linear entre as colunas (Usando gráfico do Seaborn por exemplo) *3
  - Gerar scatter entre outras opções para análise gráfica
  - Etc...
- Utilizar Botstrap pode ajudar a criar HTML mais amigável e responsivo
- Projeto inspirado no Pandas Profiling:
  https://github.com/ydataai/ydata-profiling
- O Kaggle é outra fonte de inspiração. O Kaggle realizar análise automatizada das colunas dos data sets *1
- O Explore do HVPLOT também é uma inspiração para este projeto *2
- Podemos testar o projeto com os dados do Vegas Dataset:
  https://vega.github.io/vega-datasets/
- Podemos processar qualquer data frame, importado via CSV, TSV, planilhas, etc...
- Utilizar LLMs (Large Language Models - IA Generativa) para enriquecer o projeto (opcional)
  - Utilizar tecnologia de IA Copilot ou similares para ajudar no desenvolvimento (opcional)
  - Explicar como foi usado Copilot e/ou IA Generativa no projeto (se utilizar LLMs ou Copilot)
- Documentar o projeto com MARKDOWN explicando as diferentes análises feitas
- Analisar os dados de forma visual com estatística descritiva
- Atenção na qualidade, organização, sequência e coerência do relatório em Jupyter notebook e do HTML gerado
- Teste: Pedir para alguém, que não sabe sobre o projeto, voluntáriamente revisar e fornecer dicas e *insights* para refinamento da análise de dados automatizada
  - Lembre-se do Hallway Testing...
- Entrega
  -  Publicar o Jupyter notebook do GitHub
  -  Enviar o link do repo do projeto na tarefa do MS Teams
  -  Apenas um aluno da equipe precisa entregar o projeto, lembre-se de citar os nomes dos participantes no Jupyter de análise
- Atenção: Não é para usar o Pandas Profiling para gerar a análise! A equipe deve criar sua ferramenta de análise
