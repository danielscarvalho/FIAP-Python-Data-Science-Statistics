### Introdução

Alguns dos principais recursos que o scikit-learn fornece: Ele pressupõe um conhecimento básico de práticas de aprendizado de máquina (ajuste de modelo, previsão, validação cruzada, etc.). 

O Scikit-learn é uma biblioteca de aprendizado de máquina (ML - Machine Learning) de código aberto que oferece suporte ao aprendizado supervisionado e não supervisionado. Ele também fornece várias ferramentas para ajuste de modelo, pré-processamento de dados, seleção de modelo, avaliação de modelo e muitos outros utilitários.

Ajuste e previsão: noções básicas do estimador

O Scikit-learn fornece dezenas de algoritmos e modelos de aprendizado de máquina integrados, chamados estimadores (estimators). Cada estimador pode ser ajustado a alguns dados usando seu método de ajuste (fit).

O método fit geralmente aceita 2 entradas:

A matriz de amostras (ou matriz de design) X. O tamanho de X é tipicamente (n_samples, n_features), o que significa que as amostras são representadas como linhas (registros) e os recursos são representados como colunas (campos, features).

Os valores-alvo y que são números reais para tarefas de regressão, ou inteiros para classificação (ou qualquer outro conjunto discreto de valores). Para tarefas de aprendizado não supervisionado, y não precisa ser especificado. y é geralmente uma matriz 1d onde a i-nésima entrada corresponde ao alvo da i-nésima amostra (linha) de X.

Normalmente, espera-se que X e y sejam matrizes numpy ou tipos de dados equivalentes semelhantes a matrizes, embora alguns estimadores funcionem com outros formatos, como matrizes esparsas.

Uma vez que o estimador é ajustado, pode ser usado para prever valores-alvo (target) de novos dados. Você não precisa treinar novamente o estimador

Variable Notes

pclass: A proxy for socio-economic status (SES)

1st = Upper
2nd = Middle
3rd = Lower

age: Age is fractional if less than 1. If the age is estimated, is it in the form of xx.5

sibsp: The dataset defines family relations in this way...
Sibling = brother, sister, stepbrother, stepsister
Spouse = husband, wife (mistresses and fiancés were ignored)

parch: The dataset defines family relations in this way...
Parent = mother, father
Child = daughter, son, stepdaughter, stepson
Some children travelled only with a nanny, therefore parch=0 for them.

````python
titanic_df = titanic_df.dropna(subset=["Age"])
````

````python
titanic_df.sample(10)
````

````python
titanic_df["SexNumber"] = titanic_df['Sex'].map({'male': 1, 'female': -1})
````

````python
titanic_df.sample(10)
````

````python
x = titanic_df[["Pclass","SexNumber","Age"]]
y = titanic_df["Survived"]
````

````python
ann_model = MLPClassifier(max_iter = 1000)
````

````python
ann_model.fit(x, y)
````

````python
ann_model.score(x, y)
````

````python
predict_dict = {"Pclass": 1,
                "SexNumber": -1,
                "Age": 22}

ann_model.predict(pd.DataFrame(predict_dict, index=[1]))
````

````python
ann_model.predict(np.array([[1, 1, 2]]))
````

````python
ann_model.predict(titanic_df[["Pclass","SexNumber","Age"]])
````

````python
titanic_df["dead_or_alive"] = ann_model.predict(titanic_df[["Pclass","SexNumber","Age"]])
````

### Bonus

Linear Regression Model do scikit-learn... usando os mesmos dados do Titanic...

````python
# Step 1 - Import the library scikit-learn linear regression model:
from sklearn.linear_model import LinearRegression

# Step 2 - Create an instance of the model:
lr_model = LinearRegression()

# Step 3 - Train the model:
lr_model.fit( titanic_df[["Pclass","SexNumber","Age"]],
              titanic_df["Survived"])
````

````python
lr_model.score(titanic_df[["Pclass","SexNumber","Age"]],
              titanic_df["Survived"])
````

````python
lr_model.coef_
````

````python
# Step 4 - Check model:

titanic_lrm_df = titanic_df[["Pclass","SexNumber","Age","dead_or_alive"]]

titanic_lrm_df["lr_prediction"] = lr_model.predict(titanic_df[["Pclass","SexNumber","Age"]])

titanic_lrm_df.sample(10)
````
