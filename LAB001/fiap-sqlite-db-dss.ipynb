```python
import sqlite3

con = sqlite3.connect("chinook.db")
```

```python
cur = con.cursor()
```

```python
res = cur.execute("select * from albums")

for registro in res:
    print(registro)
```

```python
for record in cur.execute("select date(), time(), sqlite_version()"):
    print(record)
```

```python
con.close()
```
