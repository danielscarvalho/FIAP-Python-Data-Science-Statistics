Importar as LIBs e os dados para um DataFrame Pandas

````python
import pandas as pd
import numpy as np

bra_df = pd.read_csv("brasileirao_serie_a.csv", encoding="utf-8")

#utf-8
#iso-8859-1
#ascii
````

Vamos espiar os dados...

````python
bra_df
````

Note o tipo do objeto/variável:

````python
type(bra_df)
````

Vamos criar uma nova coluna com a soma dos gols dos times visitantes + mandantes

````python
bra_df['gols_total'] = bra_df['gols_mandante'] + bra_df['gols_visitante']
````

Qual foi a maior goleada que temos registrado nestes dados?

````python
bra_df['gols_total'].max()
````

Qual é a média de gols nos jogos?

````python
bra_df['gols_total'].mean()
````

Vamos ver a distribuição dos dados de gols totais por meio de um histograma:

````python
bra_df['gols_total'].hist()
````

Agora vamos verificar quantos gols temos em média, em uma amostra aleatória de 100 jogos:  

````python
bra_df['gols_total'].sample(100).mean()
````

Quantos jogos temos neste DataFrame para análise?

````python
bra_df['gols_total'].count()
````

10% desta população:

````python
8078*.1
````

Agora vamos verificar a média de gols com uma amostra aleatória de 10% dos jogos

````python
bra_df['gols_total'].sample(807).mean()
````

E agora visualizar o histograma, podemos ajustar o tamanho da amostra aleatória dos jogos...

````python
import math

bra_df['gols_total'].sample(math.floor(8078*.4)).hist()
````

E como é a distribuição dos gols em toda a amostra de dados...

````python
bra_df['gols_total'].hist()
````
