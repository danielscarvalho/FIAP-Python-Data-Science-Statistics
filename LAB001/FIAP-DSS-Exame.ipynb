> Data de entrega: 12 de Dezembro de 2024 (Tarefa no MS Teams, link para o repo no GitHub com o Jupyter notebook do projeto)

Descrição do Projeto de Análise de Dados:

  - Projeto individual
  - Análise de dados do "Brasileirão Série A" (Na pasta data deste repo no GitHub)
      - [data/brasileirao_serie_a.csv.gz](data/brasileirao_serie_a.csv.gz)
  - Organizar e realizar análise de dados conforme as etapas de ciências dos dados que trabalhamos:
      1. Ask - Estabelecer questões de pesquisa, definir escopo
      2. Get - Obter dados de fontes confiáveis
      3. Explore - Combinar dados, limpar, revisitar questões
      4. Model - Experimentar diferentes modelos de IA (ML) com Scikit-learn e estatística descritiva
      5. Communicate - Finalizar relatório com todo o processo de análise, avançar na qualidade dos gráficos texto (MARKDOWN), e testar com usuário não envolvido no projeto para obter _insights_
  - Elaborar 10 questões de pesquisa sobre os dados (escopo) a serem respondidas/encontradas, questões avançadas
  - Importar os dados em um data frame Pandas
  - Realizar a limpeza de dados...
  - Antes de cada célula de código, explicar o que vai ser feito (MARKDOWN)
  - Depois de cada célula de output com gráfico, tabela, valores, explicar o que pode ser observados nos dados/visualização (Análise de dados)
  - Lembre-se: Alguém que não programa deve ser capaz de ler seu notebook de análise e entender o que está ocorrendo (MARKDOWN, Gráficos, tabelas, etc...)
  - Depois das células de código ou output explicar o que o código mostra, gráfico, tabela, valores, etc... (MARKDOWN)
  - Realizar análise de dados com estatística descritiva
  - Para modelagem de ML (Machine Learning - IA) utilizar e comparar os resultados de no mínimo 2 algoritmos, um de **regressão** e outro de **classificação** do Scikit-learn, ML supervisionada. Explicar o label/tag utilizado na aprendizagem (fit) do seu modelo
  - Gerar visualização para análise dos resutlados do modelo de ML (IA)
  - Separe os dados em 70% para treinamento e o demais para testes
  - Explicar os passos de limpeza dos dados no data frame Pandas
  - Analisar os dados de forma visual com estatística descritiva
  - Usar visualização histograma, scatter, correlação entre outros
  - Usar gráficos do Seaborn (Python LIB para gráficos)
  - Criar visualizações INCRÍVEIS (avançadas) com Seaborn, Hvplot, etc...
  - Atenção na qualidade, organização, sequência e coerência do relatório em Jupyter notebook
  - Finalizar com uma conclusão geral da análise
  - Pedir para alguém, que não sabe sobre o projeto, voluntáriamente revisar e fornecer dicas e *insights* para refinamento da análise de dados
  - Lembre-se do Hallway Testing...
  - Entrega
