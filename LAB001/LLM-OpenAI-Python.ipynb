````python
from openai import OpenAI

client = OpenAI()
````

````python
completion = client.chat.completions.create(
    temperature=.7,
    model="gpt-4o",
    messages=[
        {"role": "user", "content": "Escreva um breve poema sobre IA"}
    ]
)
````

````python
type(completion)
````

````python
completion.choices[0].message
````

````python
print(completion.choices[0].message.content)
````

````python
from IPython.display import display_markdown

def queryAi(query):
    completion = client.chat.completions.create(
        temperature=.7,
        model="gpt-4o",
        messages=[
            {"role": "user", "content": query}
        ]
    )
    display_markdown(completion.choices[0].message.content, raw=True)
````

````python
queryAi('Exemplos de datasets classicos usados em ciências dos dados')
````

````python
queryAi('Explique os cursos universitários de: Engenharia de Software, Ciências da Computação e Engenharia de Computação')
````

````python
queryAi('Criar uma função para calcular o número de fibonacci, em Python, C e Java')
````

````python
def fibonacci(n):
    if n <= 0:
        return 0
    elif n == 1:
        return 1
    else:
        a, b = 0, 1
        for _ in range(2, n + 1):
            a, b = b, a + b
        return b

# Exemplo de uso
print(fibonacci(10))  # Saída: 55
````
