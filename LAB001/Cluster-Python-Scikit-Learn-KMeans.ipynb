A inércia pode ser reconhecida como uma medida de quão internamente coerentes são os clusters. Ela sofre de várias desvantagens:

- A inércia faz a suposição de que os clusters são convexos e isotrópicos, o que nem sempre é o caso. Ela responde mal a clusters alongados ou variedades com formas irregulares.

- A inércia não é uma métrica normalizada: sabemos apenas que valores mais baixos são melhores e zero é o ideal. Mas em espaços de dimensões muito altas, as distâncias euclidianas tendem a se tornar infladas (esta é uma instância da chamada "maldição da dimensionalidade"). Executar um algoritmo de redução de dimensionalidade, como a análise de componentes principais (PCA) antes do agrupamento de k-means, pode aliviar esse problema e acelerar os cálculos.

[Scikit-learn documentation]

O algoritmo KMeans agrupa dados tentando separar amostras em n grupos de variância igual, minimizando um critério conhecido como inércia ou soma dos quadrados dentro do cluster (veja abaixo). Este algoritmo requer que o número de clusters seja especificado. Ele escala bem para grandes números de amostras e tem sido usado em uma grande variedade de áreas de aplicação em muitos campos diferentes.

O algoritmo k-means divide um conjunto de N amostras X em K clusters disjuntos C, cada um descrito pela média 
das amostras no cluster. As médias são comumente chamadas de "centroides" do cluster; observe que elas não são, em geral, pontos de mi_j, embora vivam no mesmo espaço.

O algoritmo K-means visa escolher centroides que minimizem a inércia ou o critério da soma dos quadrados dentro do cluster:

$$\sum_{i=0}^{n}\min_{\mu_j \in C}(||x_i - \mu_j||^2)$$

Se tiver seus dados em um único vetor (y), podemos usar `np.arange`para gerar um tipo de índice, então fornecendo dados x (len), y;

ou se seus dados estiverem em uma matriz bidimensional (x, y). Podemos visualizar em um scatter, e depois destacar os dados em cluster (agrupamentos)...

### LAB

```python
import numpy as np
import matplotlib.pyplot as plt
```

Vamos gerar dados sintéticos matemáticos discretos para teste (sample data).<br>
Note que estes dados não tem label (tag).

- 20 números aleatórios entre 0 e 1

```python
floats = np.random.random(20)
floats
```

- 20 números aleatórios entre 30 e 50

```python
ints = np.random.randint(30,50,20)
ints
```

- 20 números aleatórios entre -200 e -100

```python
negatives = np.random.randint(-200,-100,20)
negatives
```

Criando um vetor novo com os dados sintéticos

```python
sample_data = np.concatenate(np.array([floats, ints, negatives]))
sample_data
```

Embaralhando os dados

```python
np.random.shuffle(sample_data)
sample_data
```

Agora vamos visualizar este vetor com dados sintéticos em um gráfico 2D scatter plot:

```python
plt.scatter(np.arange(len(sample_data)), # Eixo X 
                          sample_data,                 # Eixo Y
                          c = sample_data)             # Cor
```

A partir do vetor com os dados sintéticos vamos gerar uma matriz, sendo que x listamos valores sequenciais, e os dados sintéticos em y: 

```python
sample_matrix = np.array(list(zip(np.arange(len(sample_data)), # counts (len) x                      
                                                sample_data))) # values y
```

```python
sample_matrix
```

Vamos ver as dimenções das duas variávies:

```python
sample_data.shape
```

```python
sample_matrix.shape
```

Agora vamos usar IA ML não supervisionada KMeans do Scikit-learn, para realizar o cluster dos dados, ou seja, como estes dados podem ser agrupados de forma autonoma:

```python
from sklearn.cluster import KMeans

kmeans_model = KMeans(n_clusters=3, random_state=1)

x = sample_matrix

kmeans_model.fit(x)
```

O KMeans gerou 3 cluster, e podemos agora obter estes clusters para cada elemento (linha) da matriz:

```python
kmeans_model.labels_
```

```python
np.column_stack(np.unique(kmeans_model.labels_, return_counts=True))
```

```python
plt.hist(kmeans_model.labels_)
plt.show()
```

```python
kmeans_model.score(sample_matrix)
```

```python
kmeans_model.cluster_centers_
```

```python
kmeans_model.predict([[5, -130]]) # x e y
```

```python
kmeans_model.predict([[0, 22], 
                      [0, .44]]) # x e y
```

```python
from sklearn import metrics

labels = kmeans_model.labels_

metrics.silhouette_score(x, labels, metric='euclidean')
```

Visualização para verificar como o cluster foi organizado pelo algoritmo (modelo) KMeans:

```python
import matplotlib.pyplot as plt

# Step size of the mesh. Decrease to increase the quality of the VQ.
h = 0.02  # point in the mesh [x_min, x_max]x[y_min, y_max].

# Plot the decision boundary. For that, we will assign a color to each
x_min, x_max = sample_matrix[:, 0].min() - 1, sample_matrix[:, 0].max() + 1
y_min, y_max = sample_matrix[:, 1].min() - 1, sample_matrix[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

# Obtain labels for each point in mesh. Use last trained model.
Z = kmeans_model.predict(np.c_[xx.ravel(), yy.ravel()])

# Put the result into a color plot
Z = Z.reshape(xx.shape)
plt.figure(1)
plt.clf()

plt.imshow(
    Z,
    interpolation="nearest",
    extent=(xx.min(), xx.max(), yy.min(), yy.max()),
    cmap=plt.cm.Paired,
    aspect="auto",
    origin="lower",
)

plt.plot(sample_matrix[:, 0], sample_matrix[:, 1], "k.", markersize=2)

# Plot the centroids as a white X
centroids = kmeans_model.cluster_centers_

plt.scatter(
    centroids[:, 0],
    centroids[:, 1],
    marker="x",
    s=169,
    linewidths=3,
    color="w",
    zorder=10,
)

plt.title(
    "AI - ML\n"
    "Scikit-learn K-means clustering\n"
    "Centroids are marked with white cross"
)

plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)
plt.xticks(())
plt.yticks(())
plt.show()
```

Colocando os dados em um data frame Pandas, e adicionando os lables de cluster gerados pelo KMeans:

```python
import pandas as pd
```

```python
data_df = pd.DataFrame(sample_matrix, columns=["X","Y"])
data_df["cluster_label"] = kmeans_model.labels_ 
```

```python
data_df
```

Visualização do cluster com Seaborn:

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="darkgrid")

# Set up the figure
f, ax = plt.subplots(figsize=(8, 8))

# Draw a contour plot to represent each bivariate density
sns.kdeplot(
    data = data_df,
    x = "X",
    y = "Y",
    hue = "cluster_label",
    thresh = .1,
)

```

```python
import seaborn as sns

sns.set_theme(style="ticks")

sns.pairplot(data_df, hue="cluster_label")
```
